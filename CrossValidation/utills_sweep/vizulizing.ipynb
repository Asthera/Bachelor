```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('general_single.csv')

# Assuming you have the data loaded in variables x, y, and labels
x = df['Test F1_mean']
y = df['Test F1_max']
labels = df['sweep_type']

# Create a color map based on the labels
label_color_map = {
    'without augmentation': 'skyblue',
    'affine': 'darkred',
    'elastic': 'orange',
    'pixel-level': 'limegreen',
    'grid_distortion': 'purple',
    'random_erasing_small_region': 'maroon'
}

colors = [label_color_map[label] for label in labels]

# Plotting
plt.figure(figsize=(10, 6))
scatter = plt.scatter(x, y, c=colors, s=100, alpha=0.7)


# Add title and labels
plt.title("Effect of Data Augmentation Techniques on Test F-1 Scores")
plt.xlabel("Test F-1 Mean Score")
plt.ylabel("Test F-1 Max Score")

# Add grid lines
plt.grid(True)

# Create a legend
legend_labels = {
    'without augmentation': 'skyblue',
    'affine': 'darkred',
    'elastic': 'orange',
    'pixel-level': 'limegreen',
    'grid_distortion': 'purple',
    'random_erasing_small_region': 'maroon'
}
handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10) for label, color in legend_labels.items()]
plt.legend(handles, legend_labels.keys(), title="Augmentation Techniques", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
