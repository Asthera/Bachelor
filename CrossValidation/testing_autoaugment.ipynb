```python
from utils import *


orig_img = Image.open("testing_image.png")


orig_img
```

```python
import torchvision.transforms as T
import matplotlib.pyplot as plt

policies = [T.AutoAugmentPolicy.CIFAR10, T.AutoAugmentPolicy.IMAGENET, T.AutoAugmentPolicy.SVHN]
augmenters = [T.AutoAugment(policy) for policy in policies]

# Show the original image and the augmented images for each policy 5 times
for _ in range(10):
    fig, axs = plt.subplots(1, 4, figsize=(15, 5))
    axs[0].imshow(orig_img, cmap="gray")
    axs[0].set_title("Original")
    for i, (policy, augmenter) in enumerate(zip(policies, augmenters), 1):
        augmented_img = augmenter(orig_img)
        axs[i].imshow(augmented_img, cmap="gray")
        axs[i].set_title(policy)
    plt.show()
```
