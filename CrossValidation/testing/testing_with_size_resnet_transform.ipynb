```python
from PIL import Image
import torchvision
from monai.transforms import ScaleIntensityRange
```

```python
image = Image.open("testing_image.png")
image
```

```python

transform = torchvision.transforms.Compose([
    torchvision.transforms.ToTensor(),
    #TransformResize(600),
    #TransformPad(600),
    ScaleIntensityRange(a_min=0,a_max=255,b_min=0,b_max=1),
    torchvision.transforms.ToPILImage()
])
```

```python
img = torchvision.transforms.ToTensor()(image)

for i in img:
    print(i.shape)

image = transform(image)
image
```

```python
# applying default transform from resnet-18
transforms = torchvision.models.ResNet18_Weights.IMAGENET1K_V1.transforms(antialias=True)
image_tensor = torchvision.transforms.ToTensor()(image)
print(image_tensor.shape) 

# make from 1 channel to 3 channel
if image_tensor.shape[0] == 1:
    image_tensor = image_tensor.repeat(3, 1, 1)

# show the image
image = torchvision.transforms.ToPILImage()(image_tensor)
image.show()

# apply the transforms
image =  transforms(image_tensor)
```
